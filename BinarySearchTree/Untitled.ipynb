```python
class BST:
    
    def __init__(self, value):
        self.key = value
        self.left = None
        self.right = None

    def insert(self, data):
        
        #if tree is empty, the new node will be the root node
        # root was declared as BST(None)
        if self.key is None:  
            self.key = data
            return
        
        #if tree is not empty- check the value of the new node and accordingly
        #decide if it needs to be placed in Left ST or Right ST
        
        if self.key > data:  # check if root is higher that data
            
            #if left subtree is empty then create a new node, if not call the insert function recursively till you enter the new node
            
            if self.left:            # left child is not None
                self.left.insert(data)
                
            else:                    # left child is None
                self.left = BST(data)
                
        
        else:              # data is greater than root - hence go to right subtree
              
            if self.right:
                self.right.insert(data)
            else:
                self.right = BST(data)
                
                
    def search(self, data):
        
        if self.key == data:     # check if data is equal to the root/called node
            print("Node found!")
            return
        
        if self.key > data:   # search in left sub tree
            if self.left is None:
                print("Node is not present in the tree") 
            else:
                self.left.search(data)
                
        else:                    # search in right subtree
            if self.right:
                self.right.search(data)
            else:
                print("Node is not present in the tree")
                

    def preorder(self):
        
        print(self.key,"-->", end = "")
        if self.left:
            self.left.preorder()
        if self.right:
            self.right.preorder()
            
    def inorder(self):
        
        if self.left:
            self.left.inorder()
        print(self.key,"-->", end = "")
        
        if self.right:
            self.right.inorder()
        
    
            
    def postorder(self):
        
        
        if self.left:
            self.left.postorder()
        if self.right:
            self.right.postorder()
        
        print(self.key,"-->", end = "")
        
    
    
    def delete(self, data):
        
        
        if self.key == None:       # if root is emtpy then return
            print("Tree is empty! Can't delete node")
            return
        
        if data < self.key:      # node that is to be deleted is in left subtree
            
            if self.left:
                self.left = self.left.delete(data)
            else:
                print("Given node is not present in the tree")
        elif data > self.key:    #node that is to be deleted in the right subtree
            if self.right:
                self.right = self.right.delete(data)
            else:
                print("Given node is not present in the tree")
                
        else:                     #node that is to be deleted is the root/current node
                
            if self.left is None:   #if the node that is to be deleted has only one child
                temp = self.right
                self = None

                return temp

            if self.right is None:   #if the node that is to be deleted has only one child

                temp = self.left
                self = None
                return temp

           # if the node that is to be deleted has two childs 
            node  = self.right     # because we wante to replace the current node with the smallest child from the rightsubtree

            while node.left:      # smallest value will be present in current node's left subtree
                node=node.left

            self.key = node.key

            self.right = self.right.delete(node.key)
                
        return self
    
    
    def deleteroot(self, data, curr):
        
        
        if self.key == None:       # if root is emtpy then return
            print("Tree is empty! Can't delete node")
            return
        
        if data < self.key:      # node that is to be deleted is in left subtree
            
            if self.left:
                self.left = self.left.deleteroot(data, curr)
            else:
                print("Given node is not present in the tree")
        elif data > self.key:    #node that is to be deleted in the right subtree
            if self.right:
                self.right = self.right.deleteroot(data,curr)
            else:
                print("Given node is not present in the tree")
                
        else:                     #node that is to be deleted is the root/current node
                
            if self.left is None:   #if the node that is to be deleted has only one child
                temp = self.right
                
                if data == curr:
                    self.key = temp.key
                    self.left = temp.left
                    self.right = temp.right
                    temp = None
                    return
                
                self = None
                return temp

            if self.right is None:   #if the node that is to be deleted has only one child

                temp = self.left
                
                if data== curr:
                    self.key = temp.key
                    self.left = temp.left
                    self.right = temp.right
                    temp = None
                    return
                
                self = None
                return temp

           # if the node that is to be deleted has two childs 
            node  = self.right     # because we wante to replace the current node with the smallest child from the rightsubtree

            while node.left:      # smallest value will be present in current node's left subtree
                node=node.left

            self.key = node.key

            self.right = self.right.deleteroot(node.key, curr)
                
        return self
    
    

    def findmin(self):
        
        curr = self
        
        while curr.left:
            curr = curr.left
            
        print("Node with smallest key is :", curr.key)
        
    
    def findmax(self):
        
        curr = self
        
        while curr.right:
            curr = curr.right
            
        print("Node with smallest key is: ", curr.key)
    
       
# count is a helper function, not a method
def count(node):
    if node is None:
        return 0
    
    return 1+ count(node.left)+count(node.right)

root =  BST(10)

nodelist = [2,4,17,3,91,11,98]

for node in nodelist:
    root.insert(node)
            

            
       
            
        
        
```

```python
print("Preorder")
root.preorder()
print()
print("Postorder")
root.postorder()
print()
print("In-Order")
root.inorder()
print()
print(count(root))
```
